```python
import os
from PIL import ImageEnhance,Image

ORIGIN_PATH = r"/Users/maithilipawde/Desktop/ChestCancerDetection/dataset/normal"
DESTIN_PATH = r"/Users/maithilipawde/Desktop/ChestCancerDetection/preprocessed_dataset/normal"

# Create destination directory if it doesn't exist
if not os.path.exists(DESTIN_PATH):
    os.makedirs(DESTIN_PATH)

# Loop through all files in the origin directory
for filename in os.listdir(ORIGIN_PATH):
    # Construct full paths for the original and destination images
    origin_image_path = os.path.join(ORIGIN_PATH, filename)
    destin_image_path = os.path.join(DESTIN_PATH, filename)
    
    # Open the image file
    with Image.open(origin_image_path) as img:
        # Resize the image
        out = img.resize((300, 300))
        # Save the converted image to the destination directory
        out.save(destin_image_path, format="PNG")
```

```python
import os

from tensorflow.keras.preprocessing import image
import numpy as np

# Define the directory containing your dataset
dataset_dir = r'C:\Users\diwas\Desktop\deeplearningProject\preprocessed_dataset\normal'

# Define the directory where augmented images will be saved
augmented_dir = r'C:\Users\diwas\Desktop\deeplearningProject\preprocessed_dataset\augmented_normal'

# Create the ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=40,
                horizontal_flip=True,
                width_shift_range=0.2,
                height_shift_range=0.2,
                shear_range=0.2,
                zoom_range=0.2,
                )

# Create a list of all image filenames in the dataset directory
filenames = os.listdir(dataset_dir)
# Ensure the augmented directory exists
os.makedirs(augmented_dir, exist_ok=True)

# Perform data augmentation and save augmented images
for filename in filenames:
    img = image.load_img(os.path.join(dataset_dir, filename))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    # Generate augmented images
    i = 0
    for batch in datagen.flow(x, batch_size=1, save_to_dir=augmented_dir, save_prefix='augmented_' + filename[:-4], save_format='jpg'):
        i += 1
        if i > 4:  # Number of augmented images per original image
            break  # Break the loop to avoid infinite loop
```
